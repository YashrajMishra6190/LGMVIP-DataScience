# Data Cleaning

```python
def find_dirty_values(data):
    dtypes = pd.DataFrame(data.dtypes,columns=["Data Type"])
    dtypes["Unique Values"]=data.nunique().sort_values(ascending=True)
    dtypes["Null Values"]=data.isnull().sum()
    dtypes["% null Values"]=data.isnull().sum()/len(data)
    return dtypes.sort_values(by="Null Values" , ascending=False).style.background_gradient(cmap='YlOrRd',axis=0)
```

```python
terrorism_df
```

```python
null_data = find_dirty_values(terrorism_df)
null_data
```

```python
# Filling Values
# Target_Subtype, 
terrorism_df['Target_subtype'] = terrorism_df['Target_subtype'].fillna('others')
```

```python
# Fill NaN values in the State column as Unknown
terrorism_df['State']=terrorism_df['State'].fillna('Unknown')
```

```python
# Fill NaN values in the city column as Unknown
terrorism_df['City']=terrorism_df['City'].fillna('Unknown')
```

```python
# To make the Data Visualization simpler replace it with Vehicle
terrorism_df.replace( {"Vehicle (not to include vehicle-borne explosives, i.e., car or truck bombs)": "Vehicle"},inplace = True)
```

# Data Visualization - Hot zone Visualization

```python
# Number of attacks per year
plt.figure(figsize=(20,5))
plt.xticks(rotation = 90)
plt.title('Attacks by Year')
sns.countplot(x=terrorism_df.Year)
```

```python
# Weapon used for most number of attacks
plt.figure(figsize=(20,5));
sns.countplot(x=terrorism_df.Weapon_type);
```

```python
# Number of Attacks by Days of a Month.
plt.figure(figsize=(20,5));
plt.title('Attacks by Day')
sns.countplot(x=terrorism_df.Day);
```

```python
# Number of Attacks by Month.
plt.figure(figsize=(20,5));
plt.title('Attacks by Month')
sns.countplot(x=terrorism_df.Month);
```

```python
# Number of Attacks in Region
plt.figure(figsize=(28,5));
plt.rcParams.update({'font.size':10});
plt.xticks(rotation = 0)
plt.title('Attacks by Region')
sns.countplot(x=terrorism_df.Region);
```

```python
# Number of Attacks based on the Target Types.
plt.figure(figsize=(20,5));
plt.rcParams.update({'font.size':10});
plt.xticks(rotation = 90)
plt.title('Based on Target type')
sns.countplot(x=terrorism_df.Target_type);
```

```python
# Number of Attacks based on Attack Types.
plt.figure(figsize=(25,5));
plt.rcParams.update({'font.size':12});
plt.xticks(rotation = 0)
plt.title('Based on Attack type')
sns.countplot(x=terrorism_df.Attack_type);
```
